```python
import random
import time
import matplotlib.pyplot as plt


def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]
        merge_sort(left_half)
        merge_sort(right_half)
        i = j = k = 0
        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1
        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1
        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1

def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[0]
    left_half = [x for x in arr[1:] if x <= pivot]
    right_half = [x for x in arr[1:] if x > pivot]
    return quick_sort(left_half) + [pivot] + quick_sort(right_half)



def compare_algorithms():
    bubble_sort_times = []
    merge_sort_times = []
    quick_sort_times = []

    for n in range(100, 1001, 100):
        arr = [random.randint(0, 1000) for _ in range(n)]
        
        start_time = time.time()
        bubble_sort(arr)
        bubble_sort_times.append(time.time() - start_time)

        start_time = time.time()
        merge_sort(arr)
        merge_sort_times.append(time.time() - start_time)

        start_time = time.time()
        quick_sort(arr)
        quick_sort_times.append(time.time() - start_time)

    plt.plot(range(100, 1001, 100), bubble_sort_times, label='Bubble Sort')
    plt.plot(range(100, 1001, 100), merge_sort_times, label='Merge Sort')
    plt.plot(range(100, 1001, 100), quick_sort_times, label='Quick Sort')
    plt.xlabel('Array Size')
    plt.ylabel('Running Time (Seconds)')
    plt.title('Sorting Algorithms Comparison')
    plt.legend()
    plt.show()


compare_algorithms()
```
